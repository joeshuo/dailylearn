```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import os
import warnings
import platform

# --- 1. ÂÖ®Â±ÄÈÖçÁΩÆ‰∏éÁéØÂ¢ÉËÆæÁΩÆ ---
warnings.filterwarnings('ignore')

# Ëá™Âä®ÈÖçÁΩÆÂ≠ó‰ΩìÔºåÈÄÇÈÖç Windows/Mac/Linux
system_name = platform.system()
if system_name == 'Windows':
    plt.rcParams['font.sans-serif'] = ['SimHei']
elif system_name == 'Darwin':  # Mac
    plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
else:
    plt.rcParams['font.sans-serif'] = ['DejaVu Sans'] # Linux fallback
plt.rcParams['axes.unicode_minus'] = False

SEASON_1 = 'Â∫¶Â§è'

# Â≠£ËäÇÈÖçÁΩÆÂ≠óÂÖ∏ (Êñπ‰æø‰øÆÊîπÂπ¥‰ªΩ)
SEASON_CONFIG = {
    'summer': {
        'anchor': ('2025-04-01', '2025-05-15'),  # Âü∫ÂáÜÊúü
        'focus':  ('2025-06-15', '2025-09-30'),  # Ê†∏ÂøÉÂàÜÊûêÊúü
        'temp_color': '#2ca02c',                 # ÁªøËâ≤
        'load_color': '#d62728',                 # Á∫¢Ëâ≤
        'label': 'Âà∂ÂÜ∑ÁîµÈáè',
        'desc': 'È™§ÂÜ∑ÂõûË∞É'
    },
    'winter': {
        'anchor': ('2024-10-01', '2024-11-15'),
        'focus':  ('2024-12-01', '2025-02-28'),
        'temp_color': '#1f77b4',                 # ËìùËâ≤
        'load_color': '#ff7f0e',                 # Ê©ôËâ≤
        'label': 'ÂèñÊöñÁîµÈáè',
        'desc': 'ÂõûÊöñÂõûË∞É'
    }
}

def load_data(file_path):
    """ËØªÂèñÂπ∂Ê∏ÖÊ¥óÊï∞ÊçÆ"""
    try:
        # Â∞ùËØï‰∏çÂêåÁºñÁ†Å
        try:
            df = pd.read_csv(file_path, encoding='gbk')
        except UnicodeDecodeError:
            df = pd.read_csv(file_path, encoding='utf-8')
            
        # Áªü‰∏ÄÂàóÂêçÊ†ºÂºèÔºàÂéªÈô§Á©∫Ê†ºÔºâ
        df.columns = df.columns.str.strip()
        
        # ËΩ¨Êç¢Êó•Êúü
        df['date'] = pd.to_datetime(df['date'])
        df = df.sort_values('date').reset_index(drop=True)
        
        # Á°Æ‰øùÊï∞ÂÄºÂàó‰∏∫ÊµÆÁÇπÂûã
        cols_to_numeric = ['load_MWh', 'temp_mean']
        for col in cols_to_numeric:
            if col in df.columns:
                df[col] = pd.to_numeric(df[col], errors='coerce')
                
        return df
    except Exception as e:
        print(f"‚ùå ËØªÂèñÊñá‰ª∂Â§±Ë¥•: {e}")
        return None

def auto_detect_events(df, load_col='climate_load', ratio_col='ratio'):
    """
    „ÄêÊô∫ËÉΩ‰æ¶Êé¢„ÄëËá™Âä®ÂØªÊâæÂÖ≥ÈîÆÊó•Êúü
    """
    events = {}
    if df.empty: return events
    
    # 1. ÂØªÊâæÂ≥∞ÂÄº (Âç†ÊØîÊúÄÈ´òÁÇπ)
    max_idx = df[ratio_col].idxmax()
    
    events['peak'] = {
        'date': df.loc[max_idx, 'date'],
        'ratio': df.loc[max_idx, ratio_col],
        'load': df.loc[max_idx, load_col],
        'total_load': df.loc[max_idx, 'load_MWh'],
        'base_load': df.loc[max_idx, 'fixed_base']
    }

    # 2. ÂØªÊâæ‚ÄúÊ∑±Ë∞∑‚Äù (Â≥∞ÂÄºÊâÄÂú®ÊúàÂÜÖÁöÑÊúÄ‰ΩéÂç†ÊØî)
    peak_month = events['peak']['date'].month
    df_core = df[df['date'].dt.month == peak_month]
    
    if not df_core.empty:
        min_idx = df_core[ratio_col].idxmin()
        events['valley'] = {
            'date': df.loc[min_idx, 'date'],
            'ratio': df.loc[min_idx, ratio_col]
        }
    else:
        events['valley'] = None

    return events

def analyze_universal_climate_load(file_path, target_names, output_folder, mode='summer'):
    """
    „ÄêÂ§èÂÜ¨ÈÄöÁî®ÂàÜÊûêÂô®„ÄëÊ†∏ÂøÉÈÄªËæë
    """
    if mode not in SEASON_CONFIG:
        print(f"‚ùå Ê®°ÂºèÈîôËØØ: {mode}Ôºå‰ªÖÊîØÊåÅ 'summer' Êàñ 'winter'")
        return

    # Ëé∑ÂèñÈÖçÁΩÆ
    cfg = SEASON_CONFIG[mode]
    if not os.path.exists(output_folder): os.makedirs(output_folder)
    
    print(f"üîÑ ÂêØÂä®ÂàÜÊûêÊ®°Âºè: {mode.upper()} | ÁõÆÊ†áË°å‰∏öÊï∞: {len(target_names)}")
    
    df_raw = load_data(file_path)
    if df_raw is None: return

    for target in target_names:
        # Á≠õÈÄâË°å‰∏ö
        df = df_raw[df_raw['Ë°å‰∏öÂêçÁß∞'].astype(str).str.strip() == target].copy()
        if len(df) == 0:
            print(f"‚ö†Ô∏è Êú™ÊâæÂà∞Ë°å‰∏öÊï∞ÊçÆ: {target}")
            continue

        # --- 1. Áâ©ÁêÜÂü∫Á∫øÈîöÂÆö ---
        anchor_mask = (df['date'] >= cfg['anchor'][0]) & (df['date'] <= cfg['anchor'][1])
        df_anchor = df[anchor_mask]
        
        if len(df_anchor) > 5: # Ëá≥Â∞ëË¶ÅÊúâÂá†Â§©Êï∞ÊçÆÊâçÁÆóÊúâÊïà
            base_weekday = df_anchor[df_anchor['date'].dt.dayofweek <= 4]['load_MWh'].mean()
            base_weekend = df_anchor[df_anchor['date'].dt.dayofweek >= 5]['load_MWh'].mean()
            # Â¶ÇÊûúÂë®Êú´Êï∞ÊçÆÁº∫Â§±ÔºåÁî®Â∑•‰ΩúÊó•‰ª£Êõø
            if pd.isna(base_weekend): base_weekend = base_weekday
        else:
            print(f"‚ö†Ô∏è [{target}] Áº∫Â∞ëÂü∫ÂáÜÊúüÊï∞ÊçÆÔºå‰ΩøÁî®ÂÖ®Â±ÄÊúÄÂ∞èÂÄºÊõø‰ª£„ÄÇ")
            base_weekday = df['load_MWh'].min()
            base_weekend = base_weekday

        print(f"üìä [{target}] Âü∫Á∫øÈîÅÂÆö: Â∑•‰ΩúÊó• {base_weekday:.0f} / Âë®Êú´ {base_weekend:.0f}")

        # --- 2. ÂêëÈáèÂåñËÆ°ÁÆó (ÈÄüÂ∫¶‰ºòÂåñ) ---
        # Ê†áËÆ∞Âë®Êú´
        df['is_weekend'] = df['date'].dt.dayofweek >= 5
        # ÂåπÈÖçÂü∫Á∫ø
        df['fixed_base'] = np.where(df['is_weekend'], base_weekend, base_weekday)
        
        # ËÆ°ÁÆóÊ∞îË±°ÁîµÈáè (clipÈò≤Ê≠¢Ë¥üÊï∞)
        df['climate_load'] = (df['load_MWh'] - df['fixed_base']).clip(lower=0)
        
        # ËÆ°ÁÆóÂç†ÊØî (Èò≤Ê≠¢Èô§‰ª•0)
        df['ratio'] = np.where(
            df['load_MWh'] > 0, 
            (df['climate_load'] / df['load_MWh']) * 100, 
            0
        )

        # --- 3. ËÅöÁÑ¶Ê†∏ÂøÉÊó∂ÊÆµ ---
        focus_mask = (df['date'] >= cfg['focus'][0]) & (df['date'] <= cfg['focus'][1])
        df_focus = df[focus_mask].copy().reset_index(drop=True)
        
        if len(df_focus) == 0:
            print(f"‚ö†Ô∏è [{target}] Ê†∏ÂøÉÊó∂ÊÆµÊó†Êï∞ÊçÆÔºåË∑≥Ëøá„ÄÇ")
            continue

        # --- 4. Êô∫ËÉΩ‰æ¶Êé¢ ---
        events = auto_detect_events(df_focus)

        # ‰øùÂ≠òÁªìÊûú
        file_name = f"{mode}_{target}_ÂàÜÊûêÁªìÊûú.xlsx"
        df_focus.to_excel(os.path.join(output_folder, file_name), index=False)

        # --- 5. ‰∏ì‰∏öÁªòÂõæ ---
        plot_analysis(df_focus, events, target, mode, cfg, base_weekday, base_weekend, output_folder)

def plot_analysis(df, events, target, mode, cfg, base_wd, base_we, output_folder):
    """
    Áã¨Á´ãÁöÑÁªòÂõæÂáΩÊï∞ (Â∑≤‰ºòÂåñ Y ËΩ¥ËåÉÂõ¥)
    """
    # ÂàõÂª∫ÁîªÂ∏É
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True, gridspec_kw={'height_ratios': [2, 1]})
    
    # ==========================================
    # 1. ‰∏äÂõæÔºöÁîµÈáèÂ†ÜÂè†Âõæ
    # ==========================================
    ax1.set_title(f"„Äê{target}„Äë{SEASON_1}Á©∫Ë∞ÉÁîµÈáè\nÂü∫Á∫ø: Â∑•‰ΩúÊó•{base_wd:.0f} / Âë®Êú´{base_we:.0f}", fontsize=14, fontweight='bold')
    
    # ÁªòÂà∂Â†ÜÂè†Âå∫Âüü (Âü∫Á∫ø + Ê∞îË±°ÁîµÈáè)
    # Ê≥®ÊÑèÔºö‰∏∫‰∫ÜÁæéËßÇÔºåÂü∫Á∫øÈÉ®ÂàÜÁî®ÊµÖÁÅ∞Ëâ≤ÔºåÊ∞îË±°ÁîµÈáèÁî®ÈÖçÁΩÆËâ≤
    ax1.stackplot(df['date'], df['fixed_base'], df['climate_load'], 
                  labels=['Áâ©ÁêÜÂü∫Á∫ø', cfg['label']], 
                  colors=['#e0e0e0', cfg['load_color']], alpha=0.6, edgecolor='none')
    
    # ÁªòÂà∂ÊÄªÁîµÈáèÁ∫ø (Ê∑±Ëâ≤Á∫øÊù°ÂãæÂãíËΩÆÂªìÔºåÊõ¥Ê∏ÖÊô∞)
    l2, = ax1.plot(df['date'], df['load_MWh'], color='#000080', linewidth=2, label='ÊÄªÁîµÈáè')
    
    # ÁªòÂà∂Âü∫Á∫øËΩÆÂªì (ËôöÁ∫ø)
    l1, = ax1.plot(df['date'], df['fixed_base'], color='#555555', linestyle='--', linewidth=1.5, alpha=0.8)
    
    # --- üî• Ê†∏ÂøÉ‰øÆÊîπÔºöÂä®ÊÄÅËÆæÁΩÆ Y ËΩ¥ËåÉÂõ¥ üî• ---
    # ËÆ°ÁÆóÊï∞ÊçÆÁöÑÊúÄÂ∞èÂÄºÂíåÊúÄÂ§ßÂÄº
    y_min = df['load_MWh'].min()
    y_max = df['load_MWh'].max()
    y_range = y_max - y_min
    
    # ËÆæÁΩÆ‰∏ãÈôêÔºöÊúÄÂ∞èÂÄºÂáèÂéª 15% ÁöÑÊ≥¢Âä®ËåÉÂõ¥Ôºå‰ΩÜ‰∏çËÉΩÂ∞è‰∫é 0
    # ËøôÊ†∑ÂèØ‰ª•ÂéªÈô§Â∫ïÈÉ®ÁöÑÁ©∫ÁôΩÔºåÂêåÊó∂‰øùÁïô‰∏ÄÁÇπ‚ÄúÂëºÂê∏ÊÑü‚Äù
    y_bottom = max(0, y_min - y_range * 0.15)
    y_top = y_max + y_range * 0.15
    
    ax1.set_ylim(bottom=y_bottom, top=y_top)
    # ---------------------------------------

    # Ê†áÊ≥®Â≥∞ÂÄº
    if events.get('peak'):
        pk = events['peak']
        ax1.annotate(f"Â≥∞ÂÄº\n{pk['load']:.0f}", 
                     xy=(pk['date'], pk['total_load']), 
                     xytext=(0, 20), textcoords='offset points',
                     arrowprops=dict(arrowstyle='->', color='black', lw=1.5), 
                     ha='center', fontweight='bold', fontsize=10,
                     bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8)) # Âä†‰∏™ÁôΩÂ∫ïÊ°ÜÊõ¥Ê∏ÖÊô∞

    ax1.set_ylabel('Êó•Áî®ÁîµÈáè (MWh)', fontsize=11)
    ax1.grid(True, linestyle='--', alpha=0.4) # ÁΩëÊ†ºÁ∫øÊ∑°‰∏ÄÁÇπ
    
    # ÂèåËΩ¥ÁªòÂà∂Ê∞îÊ∏©
    ax1_temp = ax1.twinx()
    l3, = ax1_temp.plot(df['date'], df['temp_mean'], color=cfg['temp_color'], linestyle='-', linewidth=1.5, label='Êó•ÂùáÊ∞îÊ∏©', alpha=0.8)
    ax1_temp.set_ylabel('Ê∞îÊ∏© (¬∞C)', color=cfg['temp_color'])
    ax1_temp.tick_params(axis='y', labelcolor=cfg['temp_color'])

    # ÂêàÂπ∂Âõæ‰æã (‰ºòÂåñÁâà)
    import matplotlib.patches as mpatches
    # ÊâãÂä®ÂàõÂª∫Âõæ‰æãÂè•ÊüÑÔºåÁ°Æ‰øùÈ¢úËâ≤ÂØπÂ∫îÊ≠£Á°Æ
    legend_handles = [
        mpatches.Patch(color='#e0e0e0', label='Áâ©ÁêÜÂü∫Á∫ø(Âü∫Á°Ä)'),
        mpatches.Patch(color=cfg['load_color'], alpha=0.6, label=cfg['label']),
        l2, # ÊÄªÁîµÈáèÁ∫ø
        l3  # Ê∞îÊ∏©Á∫ø
    ]
    ax1.legend(handles=legend_handles, loc='upper left', frameon=True, ncol=4, fontsize=10)

    # ==========================================
    # 2. ‰∏ãÂõæÔºöÂç†ÊØîÂàÜÊûê
    # ==========================================
    ax2.plot(df['date'], df['ratio'], color=cfg['load_color'], linewidth=2, label=f'{cfg["label"]}Âç†ÊØî')
    ax2.fill_between(df['date'], 0, df['ratio'], color=cfg['load_color'], alpha=0.15)
    
    # Ê†áÊ≥®ÊúÄÈ´òÂç†ÊØî
    if events.get('peak'):
        pk = events['peak']
        ax2.annotate(f'Max: {pk["ratio"]:.1f}%', 
                     xy=(pk['date'], pk['ratio']), 
                     xytext=(0, 10), textcoords='offset points',
                     color=cfg['load_color'], fontweight='bold', ha='center')

    # Ê†áÊ≥®Ê∑±Ë∞∑
    if events.get('valley'):
        val = events['valley']
        if val['ratio'] < pk['ratio'] * 0.7:
            ax2.annotate(f"{cfg['desc']}\n{val['ratio']:.1f}%", 
                         xy=(val['date'], val['ratio']), 
                         xytext=(0, 15), textcoords='offset points',
                         arrowprops=dict(arrowstyle='->', color='black'), 
                         ha='center', fontsize=9)

    ax2.set_ylabel(f'{cfg["label"]}Âç†ÊØî (%)', fontsize=11)
    ax2.set_ylim(0, 105) 
    
    # Êó•ÊúüÊ†ºÂºèÂåñ
    ax2.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
    ax2.xaxis.set_major_locator(mdates.DayLocator(interval=7)) 
    plt.xticks(rotation=0)
    
    plt.tight_layout()
    
    # ‰øùÂ≠òÂõæÁâá
    save_img = os.path.join(output_folder, f"{mode}_{target}.png")
    plt.savefig(save_img, dpi=300, bbox_inches='tight')
    print(f"üìä ÂõæË°®Â∑≤‰øùÂ≠ò: {save_img}")
    plt.show()
    # plt.close()

if __name__ == "__main__":
    # --- Áî®Êà∑ÈÖçÁΩÆÂå∫ ---
    input_file = r"E:\AÊô∫ÁΩë\ÁîµÈáèÈ¢ÑÊµãÊï∞ÊçÆ\lightgbmÊ®°Âûã\133‰∏™Ë°å‰∏öÁîµÈáè‰∏éÂ§©Ê∞îÂØπÈΩêÊï∞ÊçÆ_0815.csv"
    output_dir = r"E:\AÊô∫ÁΩë\Êó•Â∏∏ÁõëÊµã\Á©∫Ë∞ÉÁîµÈáèËÆ°ÁÆó" 
    
    # ÁõÆÊ†áË°å‰∏öÂàóË°®
    targets = ["    Á¨¨‰∫å‰∫ß‰∏ö"]
    
    # ËøêË°åÂàÜÊûê
    analyze_universal_climate_load(input_file, targets, output_dir, mode='winter')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import os
import warnings
import platform

# --- 1. ÂÖ®Â±ÄÈÖçÁΩÆ‰∏éÁéØÂ¢ÉËÆæÁΩÆ ---
warnings.filterwarnings('ignore')

# Ëá™Âä®ÈÖçÁΩÆÂ≠ó‰Ωì
system_name = platform.system()
if system_name == 'Windows':
    plt.rcParams['font.sans-serif'] = ['SimHei']
elif system_name == 'Darwin':  # Mac
    plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
else:
    plt.rcParams['font.sans-serif'] = ['DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# SEASON_1 = 'Â∫¶Â§è'
SEASON_1 = 'Â∫¶ÂÜ¨'

# Â≠£ËäÇÈÖçÁΩÆÂ≠óÂÖ∏
SEASON_CONFIG = {
    'summer': {
        'anchor': ('2025-04-01', '2025-05-15'),  # Âü∫ÂáÜÊúü
        'focus':  ('2025-06-15', '2025-09-30'),  # Ê†∏ÂøÉÂàÜÊûêÊúü
        'temp_color': '#2ca02c',                 # ÁªøËâ≤
        'load_color': '#d62728',                 # Á∫¢Ëâ≤
        'label': 'Âà∂ÂÜ∑ÁîµÈáè',
        'desc': 'È™§ÂÜ∑ÂõûË∞É'
    },
    'winter': {
        'anchor': ('2025-10-11', '2025-11-16'),
        'focus':  ('2026-01-01', '2026-01-19'),
        'temp_color': '#1f77b4',                 # ËìùËâ≤
        'load_color': '#ff7f0e',                 # Ê©ôËâ≤
        'label': 'ÂèñÊöñÁîµÈáè',
        'desc': 'ÂõûÊöñÂõûË∞É'
    }
}

def load_and_merge_data(elec_file, weather_file, target_industry):
    """
    „ÄêÊ†∏ÂøÉÊîπÈÄ†„ÄëËØªÂèñÂÆΩË°®Ê†ºÂºèÁöÑÁîµÈáèÂíåÊ∞îË±°Êï∞ÊçÆÔºåÂπ∂ÂêàÂπ∂‰∏∫Ê†áÂáÜÊ†ºÂºè
    """
    print(f"Ê≠£Âú®ËØªÂèñÊï∞ÊçÆ...\nÁîµÈáèË°®: {elec_file}\nÊ∞îË±°Ë°®: {weather_file}")
    
    try:
        # 1. ËØªÂèñÁîµÈáèË°® (ÂÆΩË°®)
        df_elec_raw = pd.read_excel(elec_file)
        # Á≠õÈÄâÁõÆÊ†áË°å‰∏ö
        industry_row = df_elec_raw[df_elec_raw['Ë°å‰∏öÂêçÁß∞'] == target_industry]
        
        if industry_row.empty:
            print(f"‚ùå Êú™Âú®ÁîµÈáèË°®‰∏≠ÊâæÂà∞Ë°å‰∏ö: {target_industry}")
            return None
            
        # ÊèêÂèñÊó•ÊúüÂàó (ÂÅáËÆæÂàóÂêçÊòØ '20230101' ËøôÁßçÊ†ºÂºè)
        date_cols = [c for c in df_elec_raw.columns if str(c).isdigit() and len(str(c)) == 8]
        
        # ËΩ¨ÁΩÆ‰∏∫ÈïøË°®: date, load_MWh
        elec_series = industry_row[date_cols].iloc[0]
        df_elec = pd.DataFrame({
            'date': pd.to_datetime(elec_series.index, format='%Y%m%d'),
            'load_MWh': elec_series.values.astype(float)
        })
        
        # 2. ËØªÂèñÊ∞îË±°Ë°®
        # ÈªòËÆ§ËØªÂèñÁ¨¨‰∏ÄË°å‰Ωú‰∏∫Ë°®Â§¥
        df_weather_raw = pd.read_excel(weather_file)
        
        # Êô∫ËÉΩËØÜÂà´ÂàóÂêçÔºöÂèØËÉΩÊòØ‰∏≠Êñá 'Êó•Êúü' ‰πüÂèØËÉΩÊòØËã±Êñá 'DATE'
        # Â¶ÇÊûúÂàóÂêçÈáåÊ≤°Êúâ 'Êó•Êúü' ‰πüÊ≤°Êúâ 'DATE'ÔºåÂ∞ùËØïËØªÂèñ‰∏ã‰∏ÄË°å‰Ωú‰∏∫Ë°®Â§¥
        if 'Êó•Êúü' not in df_weather_raw.columns and 'DATE' not in df_weather_raw.columns:
             df_weather_raw = pd.read_excel(weather_file, header=1)

        # Á°ÆÂÆöÂàóÂêç
        date_col_name = 'DATE' if 'DATE' in df_weather_raw.columns else 'Êó•Êúü'
        # Ê®°Á≥äÂåπÈÖçÊ∏©Â∫¶ÂàóÔºöÊâæÂåÖÂê´ 'AVG' Êàñ 'Âπ≥Âùá' ‰∏î‰∏çÂåÖÂê´ 'ÊúÄÈ´ò'/'ÊúÄ‰Ωé' ÁöÑÂàóÔºåÊàñËÄÖÁõ¥Êé•ÊåáÂÆö
        # Ê†πÊçÆ‰Ω†Êèê‰æõÁöÑÊï∞ÊçÆÔºåÂàóÂêçÊòØ 'Âπ≥ÂùáÊ∏©Â∫¶' Êàñ 'AVG'
        if 'AVG' in df_weather_raw.columns:
            temp_col_name = 'AVG'
        elif 'Âπ≥ÂùáÊ∏©Â∫¶' in df_weather_raw.columns:
            temp_col_name = 'Âπ≥ÂùáÊ∏©Â∫¶'
        else:
            # ÂÖúÂ∫ïÔºöÂÅáËÆæÁ¨¨4ÂàóÊòØÂπ≥ÂùáÊ∏©Â∫¶ (Á¥¢Âºï3)
            temp_col_name = df_weather_raw.columns[3]

        df_weather = df_weather_raw[[date_col_name, temp_col_name]].copy()
        df_weather.columns = ['date', 'temp_mean']
        
        # --- üî• ‰øÆÂ§çÁÇπÂºÄÂßã üî• ---
        # ‰ΩøÁî® errors='coerce'ÔºåÈÅáÂà∞ "DATE" ËøôÁßçÊó†Ê≥ïËΩ¨Êç¢ÁöÑÂ≠óÁ¨¶‰∏≤ÂèòÊàê NaT
        df_weather['date'] = pd.to_datetime(df_weather['date'], errors='coerce')
        
        # Âà†Èô§Êó•ÊúüËΩ¨Êç¢Â§±Ë¥•ÁöÑË°åÔºàÂç≥Âà†ÊéâÈÇ£Ë°åËã±ÊñáË°®Â§¥Ôºâ
        df_weather = df_weather.dropna(subset=['date'])
        # --- ‰øÆÂ§çÁÇπÁªìÊùü ---

        df_weather['temp_mean'] = pd.to_numeric(df_weather['temp_mean'], errors='coerce')

        # 3. ÂêàÂπ∂Êï∞ÊçÆ
        # ‰ΩøÁî® inner join Êàñ left join (‰ª•ÁîµÈáèÊï∞ÊçÆ‰∏∫ÂáÜ)
        df_merged = pd.merge(df_elec, df_weather, on='date', how='left')
        
        # ÊéíÂ∫èÂπ∂ÈáçÁΩÆÁ¥¢Âºï
        df_merged = df_merged.sort_values('date').reset_index(drop=True)
        
        return df_merged

    except Exception as e:
        print(f"‚ùå Êï∞ÊçÆËØªÂèñÊàñÂêàÂπ∂Â§±Ë¥•: {e}")
        import traceback
        traceback.print_exc()
        return None

def auto_detect_events(df, load_col='climate_load', ratio_col='ratio'):
    """„ÄêÊô∫ËÉΩ‰æ¶Êé¢„ÄëËá™Âä®ÂØªÊâæÂÖ≥ÈîÆÊó•Êúü"""
    events = {}
    if df.empty: return events
    
    # 1. ÂØªÊâæÂ≥∞ÂÄº (Âç†ÊØîÊúÄÈ´òÁÇπ)
    max_idx = df[ratio_col].idxmax()
    
    events['peak'] = {
        'date': df.loc[max_idx, 'date'],
        'ratio': df.loc[max_idx, ratio_col],
        'load': df.loc[max_idx, load_col],
        'total_load': df.loc[max_idx, 'load_MWh'],
        'base_load': df.loc[max_idx, 'fixed_base']
    }

    # 2. ÂØªÊâæ‚ÄúÊ∑±Ë∞∑‚Äù
    peak_month = events['peak']['date'].month
    df_core = df[df['date'].dt.month == peak_month]
    
    if not df_core.empty:
        min_idx = df_core[ratio_col].idxmin()
        events['valley'] = {
            'date': df.loc[min_idx, 'date'],
            'ratio': df.loc[min_idx, ratio_col]
        }
    else:
        events['valley'] = None

    return events

def analyze_universal_climate_load(elec_file, weather_file, target_names, output_folder, mode='summer'):
    """„ÄêÂ§èÂÜ¨ÈÄöÁî®ÂàÜÊûêÂô®„ÄëÊ†∏ÂøÉÈÄªËæë"""
    if mode not in SEASON_CONFIG:
        print(f"‚ùå Ê®°ÂºèÈîôËØØ: {mode}")
        return

    cfg = SEASON_CONFIG[mode]
    if not os.path.exists(output_folder): os.makedirs(output_folder)
    
    print(f"üîÑ ÂêØÂä®ÂàÜÊûêÊ®°Âºè: {mode.upper()}")

    for target in target_names:
        # --- 1. Âä†ËΩΩÂπ∂ÂêàÂπ∂Êï∞ÊçÆ (ÈíàÂØπÊØè‰∏™Ë°å‰∏öÂçïÁã¨Âä†ËΩΩ) ---
        df = load_and_merge_data(elec_file, weather_file, target)
        if df is None: continue

        # --- 2. Áâ©ÁêÜÂü∫Á∫øÈîöÂÆö ---
        anchor_mask = (df['date'] >= cfg['anchor'][0]) & (df['date'] <= cfg['anchor'][1])
        df_anchor = df[anchor_mask]
        
        if len(df_anchor) > 5:
            base_weekday = df_anchor[df_anchor['date'].dt.dayofweek <= 4]['load_MWh'].mean()
            base_weekend = df_anchor[df_anchor['date'].dt.dayofweek >= 5]['load_MWh'].mean()
            if pd.isna(base_weekend): base_weekend = base_weekday
        else:
            print(f"‚ö†Ô∏è [{target}] Áº∫Â∞ëÂü∫ÂáÜÊúüÊï∞ÊçÆÔºå‰ΩøÁî®ÂÖ®Â±ÄÊúÄÂ∞èÂÄºÊõø‰ª£„ÄÇ")
            base_weekday = df['load_MWh'].min()
            base_weekend = base_weekday

        print(f"üìä [{target}] Âü∫Á∫øÈîÅÂÆö: Â∑•‰ΩúÊó• {base_weekday:.0f} / Âë®Êú´ {base_weekend:.0f}")

        # --- 3. ÂêëÈáèÂåñËÆ°ÁÆó ---
        df['is_weekend'] = df['date'].dt.dayofweek >= 5
        df['fixed_base'] = np.where(df['is_weekend'], base_weekend, base_weekday)
        df['climate_load'] = (df['load_MWh'] - df['fixed_base']).clip(lower=0)
        df['ratio'] = np.where(df['load_MWh'] > 0, (df['climate_load'] / df['load_MWh']) * 100, 0)

        # --- 4. ËÅöÁÑ¶Ê†∏ÂøÉÊó∂ÊÆµ ---
        focus_mask = (df['date'] >= cfg['focus'][0]) & (df['date'] <= cfg['focus'][1])
        df_focus = df[focus_mask].copy().reset_index(drop=True)
        
        if len(df_focus) == 0:
            print(f"‚ö†Ô∏è [{target}] Ê†∏ÂøÉÊó∂ÊÆµÊó†Êï∞ÊçÆÔºåË∑≥Ëøá„ÄÇ")
            continue

        # --- 5. Êô∫ËÉΩ‰æ¶Êé¢ ---
        events = auto_detect_events(df_focus)

        # ‰øùÂ≠òÁªìÊûú
        file_name = f"{mode}_{target}_ÂàÜÊûêÁªìÊûú.xlsx"
        df_focus.to_excel(os.path.join(output_folder, file_name), index=False)

        # --- 6. ‰∏ì‰∏öÁªòÂõæ ---
        plot_analysis(df_focus, events, target, mode, cfg, base_weekday, base_weekend, output_folder)

def plot_analysis(df, events, target, mode, cfg, base_wd, base_we, output_folder):
    """Áã¨Á´ãÁöÑÁªòÂõæÂáΩÊï∞ (Â∑≤‰ºòÂåñ Y ËΩ¥ËåÉÂõ¥)"""
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True, gridspec_kw={'height_ratios': [2, 1]})
    
    # === ‰∏äÂõæÔºöË¥üËç∑Â†ÜÂè†Âõæ ===
    ax1.set_title(f"„Äê{target}„Äë{SEASON_1}Á©∫Ë∞ÉÁîµÈáè\nÂü∫Á∫ø: Â∑•‰ΩúÊó•{base_wd:.0f} / Âë®Êú´{base_we:.0f}", fontsize=14, fontweight='bold')
    
    ax1.stackplot(df['date'], df['fixed_base'], df['climate_load'], 
                  labels=['Áâ©ÁêÜÂü∫Á∫ø', cfg['label']], 
                  colors=['#e0e0e0', cfg['load_color']], alpha=0.6, edgecolor='none')
    
    l2, = ax1.plot(df['date'], df['load_MWh'], color='#000080', linewidth=2, label='ÊÄªÁîµÈáè')
    l1, = ax1.plot(df['date'], df['fixed_base'], color='#555555', linestyle='--', linewidth=1.5, alpha=0.8)
    
    # Âä®ÊÄÅËÆæÁΩÆ Y ËΩ¥ËåÉÂõ¥
    y_min = df['load_MWh'].min()
    y_max = df['load_MWh'].max()
    y_range = y_max - y_min
    y_bottom = max(0, y_min - y_range * 0.15)
    y_top = y_max + y_range * 0.15
    ax1.set_ylim(bottom=y_bottom, top=y_top)

    if events.get('peak'):
        pk = events['peak']
        ax1.annotate(f"Â≥∞ÂÄº\n{pk['load']:.0f}", 
                     xy=(pk['date'], pk['total_load']), 
                     xytext=(0, 20), textcoords='offset points',
                     arrowprops=dict(arrowstyle='->', color='black', lw=1.5), 
                     ha='center', fontweight='bold', fontsize=10,
                     bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))

    ax1.set_ylabel('Êó•Áî®ÁîµÈáè (MWh)', fontsize=11)
    ax1.grid(True, linestyle='--', alpha=0.4)
    
    ax1_temp = ax1.twinx()
    l3, = ax1_temp.plot(df['date'], df['temp_mean'], color=cfg['temp_color'], linestyle='-', linewidth=1.5, label='Êó•ÂùáÊ∞îÊ∏©', alpha=0.8)
    ax1_temp.set_ylabel('Ê∞îÊ∏© (¬∞C)', color=cfg['temp_color'])
    ax1_temp.tick_params(axis='y', labelcolor=cfg['temp_color'])

    import matplotlib.patches as mpatches
    legend_handles = [
        mpatches.Patch(color='#e0e0e0', label='Áâ©ÁêÜÂü∫Á∫ø(Âü∫Á°Ä)'),
        mpatches.Patch(color=cfg['load_color'], alpha=0.6, label=cfg['label']),
        l2, l3
    ]
    ax1.legend(handles=legend_handles, loc='upper left', frameon=True, ncol=4, fontsize=10)

    # === ‰∏ãÂõæÔºöÂç†ÊØîÂàÜÊûê ===
    ax2.plot(df['date'], df['ratio'], color=cfg['load_color'], linewidth=2, label=f'{cfg["label"]}Âç†ÊØî')
    ax2.fill_between(df['date'], 0, df['ratio'], color=cfg['load_color'], alpha=0.15)
    
    if events.get('peak'):
        pk = events['peak']
        ax2.annotate(f'Max: {pk["ratio"]:.1f}%', 
                     xy=(pk['date'], pk['ratio']), 
                     xytext=(0, 10), textcoords='offset points',
                     color=cfg['load_color'], fontweight='bold', ha='center')

    if events.get('valley'):
        val = events['valley']
        if val['ratio'] < pk['ratio'] * 0.7:
            ax2.annotate(f"{cfg['desc']}\n{val['ratio']:.1f}%", 
                         xy=(val['date'], val['ratio']), 
                         xytext=(0, 15), textcoords='offset points',
                         arrowprops=dict(arrowstyle='->', color='black'), 
                         ha='center', fontsize=9)

    ax2.set_ylabel(f'{cfg["label"]}Âç†ÊØî (%)', fontsize=11)
    ax2.set_ylim(0, 105) 
    
    ax2.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
    ax2.xaxis.set_major_locator(mdates.DayLocator(interval=7)) 
    plt.xticks(rotation=0)
    
    plt.tight_layout()
    
    save_img = os.path.join(output_folder, f"{mode}_{target}.png")
    plt.savefig(save_img, dpi=300, bbox_inches='tight')
    print(f"üìä ÂõæË°®Â∑≤‰øùÂ≠ò: {save_img}")
    plt.show()

if __name__ == "__main__":
    # --- Áî®Êà∑ÈÖçÁΩÆÂå∫ ---
    elec_file = r"E:\AÊô∫ÁΩë\ÊúàÂ∫¶ÁîµÂäõÈúÄÊ±ÇÈ¢ÑÊµãÊä•Âëä\0_Êï∞ÊçÆ\ÊπñÂåóÁúÅË°å‰∏öÊó•Áî®ÁîµÈáè_26.1.19.xlsx"
    weather_file = r"E:\AÊô∫ÁΩë\ÊúàÂ∫¶ÁîµÂäõÈúÄÊ±ÇÈ¢ÑÊµãÊä•Âëä\0_Êï∞ÊçÆ\Ê∞îË±°.xlsx"
    output_dir = r"E:\AÊô∫ÁΩë\ÊúàÂ∫¶ÁîµÂäõÈúÄÊ±ÇÈ¢ÑÊµãÊä•Âëä\Á©∫Ë∞ÉÁîµÈáèËÆ°ÁÆó_1.20" 
    
    # ÁõÆÊ†áË°å‰∏öÂàóË°®
    # targets = ["    Á¨¨‰∏â‰∫ß‰∏ö"]
    targets = ["    Á¨¨‰∫å‰∫ß‰∏ö"]
    # targets = ["B„ÄÅÂüé‰π°Â±ÖÊ∞ëÁîüÊ¥ªÁî®ÁîµÂêàËÆ°"]

    # ËøêË°åÂàÜÊûê
    analyze_universal_climate_load(elec_file, weather_file, targets, output_dir, mode='winter')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import os
import warnings
import platform

# --- 1. ÂÖ®Â±ÄÈÖçÁΩÆ‰∏éÁéØÂ¢ÉËÆæÁΩÆ ---
warnings.filterwarnings('ignore')

# Ëá™Âä®ÈÖçÁΩÆÂ≠ó‰Ωì
system_name = platform.system()
if system_name == 'Windows':
    plt.rcParams['font.sans-serif'] = ['SimHei']
elif system_name == 'Darwin':  # Mac
    plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
else:
    plt.rcParams['font.sans-serif'] = ['DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# SEASON_1 = 'Â∫¶Â§è'
SEASON_1 = 'Â∫¶ÂÜ¨'

# Â≠£ËäÇÈÖçÁΩÆÂ≠óÂÖ∏
SEASON_CONFIG = {
    'summer': {
        'anchor': ('2025-04-01', '2025-05-15'),  # Âü∫ÂáÜÊúü
        'focus':  ('2025-06-15', '2025-09-30'),  # Ê†∏ÂøÉÂàÜÊûêÊúü
        'temp_color': '#2ca02c',                 # ÁªøËâ≤
        'load_color': '#d62728',                 # Á∫¢Ëâ≤
        'label': 'Âà∂ÂÜ∑ÁîµÈáè',
        'desc': 'È™§ÂÜ∑ÂõûË∞É'
    },
    'winter': {
        'anchor': ('2025-4-11', '2025-4-30'),
        'focus':  ('2026-01-01', '2026-01-19'),
        'temp_color': '#1f77b4',                 # ËìùËâ≤
        'load_color': '#ff7f0e',                 # Ê©ôËâ≤
        'label': 'ÂèñÊöñÁîµÈáè',
        'desc': 'ÂõûÊöñÂõûË∞É'
    }
}

def load_and_merge_data(elec_file, weather_file, target_industry):
    """ËØªÂèñÂÆΩË°®Ê†ºÂºèÁöÑÁîµÈáèÂíåÊ∞îË±°Êï∞ÊçÆÔºåÂπ∂ÂêàÂπ∂‰∏∫Ê†áÂáÜÊ†ºÂºè"""
    print(f"Ê≠£Âú®ËØªÂèñÊï∞ÊçÆ...\nÁîµÈáèË°®: {elec_file}\nÊ∞îË±°Ë°®: {weather_file}")
    
    try:
        # 1. ËØªÂèñÁîµÈáèË°®
        df_elec_raw = pd.read_excel(elec_file)
        industry_row = df_elec_raw[df_elec_raw['Ë°å‰∏öÂêçÁß∞'] == target_industry]
        
        if industry_row.empty:
            print(f"‚ùå Êú™Âú®ÁîµÈáèË°®‰∏≠ÊâæÂà∞Ë°å‰∏ö: {target_industry}")
            return None
            
        date_cols = [c for c in df_elec_raw.columns if str(c).isdigit() and len(str(c)) == 8]
        elec_series = industry_row[date_cols].iloc[0]
        df_elec = pd.DataFrame({
            'date': pd.to_datetime(elec_series.index, format='%Y%m%d'),
            'load_MWh': elec_series.values.astype(float)
        })
        
        # 2. ËØªÂèñÊ∞îË±°Ë°®
        df_weather_raw = pd.read_excel(weather_file)
        if 'Êó•Êúü' not in df_weather_raw.columns and 'DATE' not in df_weather_raw.columns:
             df_weather_raw = pd.read_excel(weather_file, header=1)

        date_col_name = 'DATE' if 'DATE' in df_weather_raw.columns else 'Êó•Êúü'
        if 'AVG' in df_weather_raw.columns:
            temp_col_name = 'AVG'
        elif 'Âπ≥ÂùáÊ∏©Â∫¶' in df_weather_raw.columns:
            temp_col_name = 'Âπ≥ÂùáÊ∏©Â∫¶'
        else:
            temp_col_name = df_weather_raw.columns[3]

        df_weather = df_weather_raw[[date_col_name, temp_col_name]].copy()
        df_weather.columns = ['date', 'temp_mean']
        
        df_weather['date'] = pd.to_datetime(df_weather['date'], errors='coerce')
        df_weather = df_weather.dropna(subset=['date'])
        df_weather['temp_mean'] = pd.to_numeric(df_weather['temp_mean'], errors='coerce')

        # 3. ÂêàÂπ∂Êï∞ÊçÆ
        df_merged = pd.merge(df_elec, df_weather, on='date', how='left')
        df_merged = df_merged.sort_values('date').reset_index(drop=True)
        
        return df_merged

    except Exception as e:
        print(f"‚ùå Êï∞ÊçÆËØªÂèñÊàñÂêàÂπ∂Â§±Ë¥•: {e}")
        import traceback
        traceback.print_exc()
        return None

def auto_detect_events(df, load_col='climate_load', ratio_col='ratio'):
    """„ÄêÊô∫ËÉΩ‰æ¶Êé¢„ÄëËá™Âä®ÂØªÊâæÂÖ≥ÈîÆÊó•Êúü"""
    events = {}
    if df.empty: return events
    
    max_idx = df[ratio_col].idxmax()
    
    events['peak'] = {
        'date': df.loc[max_idx, 'date'],
        'ratio': df.loc[max_idx, ratio_col],
        'load': df.loc[max_idx, load_col],
        'total_load': df.loc[max_idx, 'load_MWh'],
        'base_load': df.loc[max_idx, 'fixed_base']
    }

    peak_month = events['peak']['date'].month
    df_core = df[df['date'].dt.month == peak_month]
    
    if not df_core.empty:
        min_idx = df_core[ratio_col].idxmin()
        events['valley'] = {
            'date': df.loc[min_idx, 'date'],
            'ratio': df.loc[min_idx, ratio_col]
        }
    else:
        events['valley'] = None

    return events

def analyze_universal_climate_load(elec_file, weather_file, target_names, output_folder, mode='summer'):
    """„ÄêÂ§èÂÜ¨ÈÄöÁî®ÂàÜÊûêÂô®„ÄëÊ†∏ÂøÉÈÄªËæë"""
    if mode not in SEASON_CONFIG:
        print(f"‚ùå Ê®°ÂºèÈîôËØØ: {mode}")
        return

    cfg = SEASON_CONFIG[mode]
    if not os.path.exists(output_folder): os.makedirs(output_folder)
    
    print(f"üîÑ ÂêØÂä®ÂàÜÊûêÊ®°Âºè: {mode.upper()}")

    for target in target_names:
        # --- 1. Âä†ËΩΩÂπ∂ÂêàÂπ∂Êï∞ÊçÆ ---
        df = load_and_merge_data(elec_file, weather_file, target)
        if df is None: continue

        # --- 2. Áâ©ÁêÜÂü∫Á∫øÈîöÂÆö (7Â§©Âü∫Á∫ø‰ºòÂåñÁâà) ---
        anchor_mask = (df['date'] >= cfg['anchor'][0]) & (df['date'] <= cfg['anchor'][1])
        df_anchor = df[anchor_mask].copy()
        
        if len(df_anchor) > 7: # Ëá≥Â∞ëÊúâ‰∏ÄÂë®ÁöÑÊï∞ÊçÆ
            # ËÆ°ÁÆóÂë®‰∏ÄÂà∞Âë®Êó•ÊØèÂ§©ÁöÑÂπ≥ÂùáÂÄº (0=Âë®‰∏Ä, 6=Âë®Êó•)
            # groupby dayofweek ÂæóÂà∞‰∏Ä‰∏™ SeriesÔºåÁ¥¢ÂºïÊòØ 0-6
            daily_baseline = df_anchor.groupby(df_anchor['date'].dt.dayofweek)['load_MWh'].mean()
            
            # ÊâìÂç∞Âü∫Á∫ø‰ø°ÊÅØ
            print(f"üìä [{target}] 7Â§©Âü∫Á∫øÈîÅÂÆö:")
            weekdays = ['Âë®‰∏Ä', 'Âë®‰∫å', 'Âë®‰∏â', 'Âë®Âõõ', 'Âë®‰∫î', 'Âë®ÂÖ≠', 'Âë®Êó•']
            for i, val in daily_baseline.items():
                print(f"   {weekdays[i]}: {val:.0f}")
        else:
            print(f"‚ö†Ô∏è [{target}] Áº∫Â∞ëÂü∫ÂáÜÊúüÊï∞ÊçÆÔºå‰ΩøÁî®ÂÖ®Â±ÄÊúÄÂ∞èÂÄºÊõø‰ª£„ÄÇ")
            min_val = df['load_MWh'].min()
            daily_baseline = pd.Series([min_val]*7, index=range(7))

        # --- 3. ÂêëÈáèÂåñËÆ°ÁÆó (ÂåπÈÖç7Â§©Âü∫Á∫ø) ---
        # Â∞ÜÊó•ÊúüÂØπÂ∫îÁöÑÊòüÊúüÂá† (0-6) Êò†Â∞ÑÂà∞Âü∫Á∫øÂÄº
        # map ÂáΩÊï∞ÈùûÂ∏∏Âø´
        df['dayofweek'] = df['date'].dt.dayofweek
        df['fixed_base'] = df['dayofweek'].map(daily_baseline)
        
        df['climate_load'] = (df['load_MWh'] - df['fixed_base']).clip(lower=0)
        df['ratio'] = np.where(df['load_MWh'] > 0, (df['climate_load'] / df['load_MWh']) * 100, 0)

        # --- 4. ËÅöÁÑ¶Ê†∏ÂøÉÊó∂ÊÆµ ---
        focus_mask = (df['date'] >= cfg['focus'][0]) & (df['date'] <= cfg['focus'][1])
        df_focus = df[focus_mask].copy().reset_index(drop=True)
        
        if len(df_focus) == 0:
            print(f"‚ö†Ô∏è [{target}] Ê†∏ÂøÉÊó∂ÊÆµÊó†Êï∞ÊçÆÔºåË∑≥Ëøá„ÄÇ")
            continue

        # --- 5. Êô∫ËÉΩ‰æ¶Êé¢ ---
        events = auto_detect_events(df_focus)

        # ‰øùÂ≠òÁªìÊûú
        file_name = f"{mode}_{target}_ÂàÜÊûêÁªìÊûú.xlsx"
        df_focus.to_excel(os.path.join(output_folder, file_name), index=False)

        # --- 6. ‰∏ì‰∏öÁªòÂõæ ---
        # ‰º†ÂÖ• daily_baseline ‰æõÁªòÂõæÊ†áÈ¢ò‰ΩøÁî®
        plot_analysis(df_focus, events, target, mode, cfg, daily_baseline, output_folder)

def plot_analysis(df, events, target, mode, cfg, daily_baseline, output_folder):
    """Áã¨Á´ãÁöÑÁªòÂõæÂáΩÊï∞ (Â∑≤‰ºòÂåñ Y ËΩ¥ËåÉÂõ¥)"""
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True, gridspec_kw={'height_ratios': [2, 1]})
    
    # === ‰∏äÂõæÔºöË¥üËç∑Â†ÜÂè†Âõæ ===
    # ËÆ°ÁÆóÂü∫Á∫øÂùáÂÄºÁî®‰∫éÊ†áÈ¢òÂ±ïÁ§∫
    avg_base = daily_baseline.mean()
    ax1.set_title(f"„Äê{target}„Äë{SEASON_1}Á©∫Ë∞ÉÁîµÈáè\n7Â§©Âü∫Á∫øÂùáÂÄº: {avg_base:.0f}", fontsize=14, fontweight='bold')
    
    ax1.stackplot(df['date'], df['fixed_base'], df['climate_load'], 
                  labels=['Áâ©ÁêÜÂü∫Á∫ø', cfg['label']], 
                  colors=['#e0e0e0', cfg['load_color']], alpha=0.6, edgecolor='none')
    
    l2, = ax1.plot(df['date'], df['load_MWh'], color='#000080', linewidth=2, label='ÊÄªÁîµÈáè')
    l1, = ax1.plot(df['date'], df['fixed_base'], color='#555555', linestyle='--', linewidth=1.5, alpha=0.8)
    
    # Âä®ÊÄÅËÆæÁΩÆ Y ËΩ¥ËåÉÂõ¥
    y_min = df['load_MWh'].min()
    y_max = df['load_MWh'].max()
    y_range = y_max - y_min
    y_bottom = max(0, y_min - y_range * 0.15)
    y_top = y_max + y_range * 0.15
    ax1.set_ylim(bottom=y_bottom, top=y_top)

    if events.get('peak'):
        pk = events['peak']
        ax1.annotate(f"Â≥∞ÂÄº\n{pk['load']:.0f}", 
                     xy=(pk['date'], pk['total_load']), 
                     xytext=(0, 20), textcoords='offset points',
                     arrowprops=dict(arrowstyle='->', color='black', lw=1.5), 
                     ha='center', fontweight='bold', fontsize=10,
                     bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))

    ax1.set_ylabel('Êó•Áî®ÁîµÈáè (MWh)', fontsize=11)
    ax1.grid(True, linestyle='--', alpha=0.4)
    
    ax1_temp = ax1.twinx()
    l3, = ax1_temp.plot(df['date'], df['temp_mean'], color=cfg['temp_color'], linestyle='-', linewidth=1.5, label='Êó•ÂùáÊ∞îÊ∏©', alpha=0.8)
    ax1_temp.set_ylabel('Ê∞îÊ∏© (¬∞C)', color=cfg['temp_color'])
    ax1_temp.tick_params(axis='y', labelcolor=cfg['temp_color'])

    import matplotlib.patches as mpatches
    legend_handles = [
        mpatches.Patch(color='#e0e0e0', label='Áâ©ÁêÜÂü∫Á∫ø(7Â§©)'),
        mpatches.Patch(color=cfg['load_color'], alpha=0.6, label=cfg['label']),
        l2, l3
    ]
    ax1.legend(handles=legend_handles, loc='upper left', frameon=True, ncol=4, fontsize=10)

    # === ‰∏ãÂõæÔºöÂç†ÊØîÂàÜÊûê ===
    ax2.plot(df['date'], df['ratio'], color=cfg['load_color'], linewidth=2, label=f'{cfg["label"]}Âç†ÊØî')
    ax2.fill_between(df['date'], 0, df['ratio'], color=cfg['load_color'], alpha=0.15)
    
    if events.get('peak'):
        pk = events['peak']
        ax2.annotate(f'Max: {pk["ratio"]:.1f}%', 
                     xy=(pk['date'], pk['ratio']), 
                     xytext=(0, 10), textcoords='offset points',
                     color=cfg['load_color'], fontweight='bold', ha='center')

    if events.get('valley'):
        val = events['valley']
        if val['ratio'] < pk['ratio'] * 0.7:
            ax2.annotate(f"{cfg['desc']}\n{val['ratio']:.1f}%", 
                         xy=(val['date'], val['ratio']), 
                         xytext=(0, 15), textcoords='offset points',
                         arrowprops=dict(arrowstyle='->', color='black'), 
                         ha='center', fontsize=9)

    ax2.set_ylabel(f'{cfg["label"]}Âç†ÊØî (%)', fontsize=11)
    ax2.set_ylim(0, 105) 
    
    ax2.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
    ax2.xaxis.set_major_locator(mdates.DayLocator(interval=7)) 
    plt.xticks(rotation=0)
    
    plt.tight_layout()
    
    save_img = os.path.join(output_folder, f"{mode}_{target}.png")
    plt.savefig(save_img, dpi=300, bbox_inches='tight')
    print(f"üìä ÂõæË°®Â∑≤‰øùÂ≠ò: {save_img}")
    plt.show()

if __name__ == "__main__":
    # --- Áî®Êà∑ÈÖçÁΩÆÂå∫ ---
    elec_file = r"E:\AÊô∫ÁΩë\ÊúàÂ∫¶ÁîµÂäõÈúÄÊ±ÇÈ¢ÑÊµãÊä•Âëä\0_Êï∞ÊçÆ\ÊπñÂåóÁúÅË°å‰∏öÊó•Áî®ÁîµÈáè_26.1.19.xlsx"
    weather_file = r"E:\AÊô∫ÁΩë\ÊúàÂ∫¶ÁîµÂäõÈúÄÊ±ÇÈ¢ÑÊµãÊä•Âëä\0_Êï∞ÊçÆ\Ê∞îË±°.xlsx"
    output_dir = r"E:\AÊô∫ÁΩë\ÊúàÂ∫¶ÁîµÂäõÈúÄÊ±ÇÈ¢ÑÊµãÊä•Âëä\Á©∫Ë∞ÉÁîµÈáèËÆ°ÁÆó_1.20" 
    
    # ÁõÆÊ†áË°å‰∏öÂàóË°®
    # targets = ["    Á¨¨‰∏â‰∫ß‰∏ö"]
    targets = ["    Á¨¨‰∫å‰∫ß‰∏ö"]
    # targets = ["B„ÄÅÂüé‰π°Â±ÖÊ∞ëÁîüÊ¥ªÁî®ÁîµÂêàËÆ°"]

    # ËøêË°åÂàÜÊûê
    analyze_universal_climate_load(elec_file, weather_file, targets, output_dir, mode='winter')
```
